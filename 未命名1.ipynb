```python
def company_extract(text_list):
    config = load_config("deep_model/config_file")
    logger = get_logger("deep_model/train.log")
        # limit GPU memory
    tf_config = tf.ConfigProto()
    tf_config.gpu_options.allow_growth = True
    with open(FLAGS.map_file, "rb") as f:
        tag_to_id, id_to_tag = pickle.load(f)

    com_list = []
    tri_list = []
    print("0")
    with tf.Session(config=tf_config) as sess:
        model = create_model(sess, Model, FLAGS.ckpt_path, config, logger)
        for i in range(len(text_list)):
            print(i)
            s = text_list[i]
            result = model.evaluate_line(sess, input_from_line(s, FLAGS.max_seq_len, tag_to_id), id_to_tag)
            res = result['entities']
            com = []
            tri = []
            print(result)
            for di in res:
                if di['type'] == 'C':
                    if di['word'] not in com:
                        com.append(di['word'])
                if di['type'] == 'T':
                    if di['word'] not in tri:
                        tri.append(di['word'])
            s = ' '.join(com)
            com_list.append(s)
            s = ' '.join(tri)
            tri_list.append(s)
    result = []
    for i in range(len(text_list)):
            c_list = com_list[i].split()
            t_list = tri_list[i].split()
            count = len(c_list)
            dex = 0
            for t in range(count):
                c_list[t] = c_list[t]
                if ',' in c_list[t] or '，' in c_list[t]:
                    continue
                if len(str(c_list[t])) > 1:
                    if len(t_list) > 0:
                        s = c_list[t]+"-"+t_list[0]+"_"+text_list[i]
                    else:
                        s = c_list[t]+"_"+text_list[i]
                    dex += 1
                    result.append(s)
                    print(s)
            if dex == 0:
                N = "None_"+text_list[i]
                result.append(N)
                
    tf.reset_default_graph()
    return result
                
```
